```python
with urllib.request.urlopen(urlLink) as url:
    data = json.loads(url.read().decode())
    #print(data)
```

```python
d1 = json.dumps(data)
print(d1)
jsonDict = json.loads(d1)
```

```python
urlList = []
for eachNode in jsonDict['response']['docs']:
    text = eachNode["web_url"]
    urlList.append(text)
print(urlList)
```

```python
import urllib3
import requests
from bs4 import BeautifulSoup

urllink = "https://www.nytimes.com/2018/03/28/business/media/cambridge-analytica-advertising-award.html"

response = requests.get(urllink)
soup = BeautifulSoup(response.content, "html.parser")
print(soup)
```

```python
#defining a method that takes in the soup and returns the text content
def getTextFromSoup(soup):
    div_box = soup.find('div', attrs={'class': 'story-body story-body-1'})
    scrappedText = ""
    p_box = soup.find_all('p', attrs={'class': 'story-body-text story-content'})
    for eachP in p_box :
        scrappedText = scrappedText + eachP.get_text()
    return scrappedText
    
```

```python
import urllib3
import requests
from bs4 import BeautifulSoup

NYTText=""
for urllink in urlList:
    response = requests.get(urllink)
    soup = BeautifulSoup(response.content, "html.parser")
    textVal = getTextFromSoup(soup)
    NYTText = NYTText + textVal
```

```python
import os.path
import codecs

filename ='NYTimesTextExtractFromBeautifulSoup.txt'
fp = codecs.open(filename, "w", "utf-8")
fp.write(NYTText) 
fp.write("\n")
fp.close()
```

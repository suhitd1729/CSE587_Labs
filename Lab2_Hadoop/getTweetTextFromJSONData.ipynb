```python
import urllib.request, json 
```

```python
json_data=open('tweets.json').read()

data = json.loads(json_data)
#print(data)
```

```python
strdata = ""
for d in data:
    tweetData = d['text']  
    strdata = strdata + tweetData
    
print(strdata.encode("utf-8"))

    
```

```python
import codecs
filename ='tweetTextExtract.txt'
fp = codecs.open(filename, "w", "utf-8")
fp.write(strdata) 
fp.write("\n")
fp.close()
```

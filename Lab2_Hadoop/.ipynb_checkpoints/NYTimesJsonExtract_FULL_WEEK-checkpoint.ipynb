```python
def getUrlList(jsonDict):
    pageUrl = []
    for eachNode in jsonDict['response']['docs']:
        text = eachNode["web_url"]
        pageUrl.append(text)
    return pageUrl
    
```

```python
def getdataList(jsonurlLink):
    with urllib.request.urlopen(jsonurlLink) as url:
        data = json.loads(url.read().decode())
        return data
```

```python
urlList = []
dataList = []
jsonDictList = []
for jsonurlLink in jsonUrlList:
    data = getdataList(jsonurlLink)
    d1 = json.dumps(data)
    jsonDict = json.loads(d1)
    pageUrl = getUrlList(jsonDict)
    urlList.append(pageUrl)
    time.sleep(7)

urlList
    
    
    
    
```

```python
#defining a method that takes in the soup and returns the text content
def getTextFromSoup(soup):
    div_box = soup.find('div', attrs={'class': 'story-body story-body-1'})
    scrappedText = ""
    p_box = soup.find_all('p', attrs={'class': 'story-body-text story-content'})
    for eachP in p_box :
        scrappedText = scrappedText + eachP.get_text()
    return scrappedText
    
```

```python
import urllib3
import requests
from bs4 import BeautifulSoup

NYTText=""
for u in urlList:
    for urllink in u:
        response = requests.get(urllink)
        soup = BeautifulSoup(response.content, "html.parser")
        textVal = getTextFromSoup(soup)
        NYTText = NYTText + textVal
```

```python
import os.path
import codecs

filename ='NYTTextExtractFullWeek.txt'
fp = codecs.open(filename, "w", "utf-8")
fp.write(NYTText) 
fp.write("\n")
fp.close()
```

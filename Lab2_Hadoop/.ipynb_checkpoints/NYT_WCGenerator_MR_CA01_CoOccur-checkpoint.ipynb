```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
f = open("coOccur_part-00000")
#print(f.read())
tableData = pd.read_table("coOccur_part-00000", sep = "\t")
tableData.columns = ['word','count']
print(tableData)    
```

```python
sortedData = tableData.sort_values('count',ascending= False)
sortedData.head(10)
```

```python
array = {}
for a, x in tableData.values:
    array[a] = x

wordcloud = WordCloud(background_color='black', width=2000,
                          height=2000)
wordcloud.generate_from_frequencies(frequencies = array)
plt.figure(figsize=(20,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
